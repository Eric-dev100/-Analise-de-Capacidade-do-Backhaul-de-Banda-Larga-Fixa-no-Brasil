```python
colunas_capacidade = ['capacidade_backhaul', 'capacidade_ocupada', 'capacidade_disponivel']

for coluna in colunas_capacidade:
    df_completo[coluna] = df_completo[coluna].fillna(0)
```

```python
df_completo = df_completo.dropna(subset=['ano_atendimento'])
df_completo = df_completo.drop(columns=['nome_regiao_metropolitana', 'id_regiao_metropolitana'])
print(df_completo.columns)
```

```python
df_agrupado = df_completo.groupby(['nome_regiao', 'ano_atendimento'])['capacidade_backhaul'].sum().reset_index()
```

```python
df_agrupado = df_agrupado.sort_values(by=['nome_regiao', 'ano_atendimento'])
```

```python
df_agrupado['capacidade_acumulada'] = df_agrupado.groupby('nome_regiao')['capacidade_backhaul'].cumsum()
```

```python
df_agrupado
```

```python
plt.figure(figsize=(14, 7))

grafico1 = sns.lineplot( data=df_agrupado, x='ano_atendimento', y='capacidade_acumulada', hue='nome_regiao', marker='o', linewidth=2.5, palette='Set1')

plt.title('Crescimento Acumulado da Capacidade Total de Backhaul por Região', fontsize=16, pad=15)
plt.xlabel('Ano de Atendimento', fontsize=12)
plt.ylabel('Capacidade Acumulada', fontsize=12)
anos_unicos = sorted(df_agrupado['ano_atendimento'].unique())
plt.xticks(anos_unicos, rotation=45)
plt.legend(title='Região', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot( data=df_completo, x='tecnologia', y='capacidade_backhaul', estimator=np.sum, errorbar=None, palette='magma', hue='tecnologia' )

plt.title('Capacidade Total de Backhaul por Tipo de Tecnologia', fontsize=16, pad=15)
plt.xlabel('Tecnologia Utilizada', fontsize=12)
plt.ylabel('Capacidade Total (Soma)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_ocupacao = df_completo.groupby('nome_regiao')[['capacidade_ocupada', 'capacidade_backhaul']].sum().reset_index()
```

```python
df_ocupacao['taxa_ocupacao_perc'] = (df_ocupacao['capacidade_ocupada'] / df_ocupacao['capacidade_backhaul']) * 100
```

```python
plt.figure(figsize=(10, 6))

sns.barplot( data=df_ocupacao, x='nome_regiao', y='taxa_ocupacao_perc', palette='Reds_r', hue='nome_regiao', legend=False )

plt.axhline(y=80, color='red', linestyle='--', linewidth=2, label='Alerta de Esgotamento (80%)')

plt.title('Taxa de Ocupação da Rede de Backhaul por Região (%)', fontsize=16, pad=15)
plt.xlabel('Região', fontsize=12)
plt.ylabel('Taxa de Ocupação (%)', fontsize=12)

for index, row in enumerate(df_ocupacao['taxa_ocupacao_perc']):
    plt.text(index, row + 1, f'{row:.1f}%', color='black', ha="center", fontsize=11, fontweight='bold')

plt.ylim(0, 100)
plt.legend()
plt.tight_layout()
plt.show()
```
